```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("owid-covid-data.csv")

dimensioni = df.shape
print("Dimensioni del dataset:", dimensioni)
metadata = df.info()
```

```python
df_continenti = df.dropna(subset=["continent"])

casi_per_continente = df_continenti.groupby("continent")["total_cases"].sum()
print(casi_per_continente)

continente1 = "Europe"
continente2 = "North America"
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
print(casi_per_continente)
```

```python
if continente1 not in df_continenti["continent"].unique():
    print(f"Errore: {continente1} non è presente nel dataset.")
if continente2 not in df_continenti["continent"].unique():
    print(f"Errore: {continente2} non è presente nel dataset.")
```

```python
df_continenti.loc[:, "total_cases"] = df_continenti["total_cases"].fillna(0)
```

```python
total_cases_world = df["total_cases"].sum()
total_cases_continente1 = df_continenti[df_continenti["continent"] == continente1]["total_cases"].sum()
total_cases_continente2 = df_continenti[df_continenti["continent"] == continente2]["total_cases"].sum()
```

```python
massimo_continente1 = df_continenti[df_continenti["continent"] == continente1]["total_cases"].max()
massimo_continente2 = df_continenti[df_continenti["continent"] == continente2]["total_cases"].max()
media_continente1 = df_continenti[df_continenti["continent"] == continente1]["total_cases"].mean()
media_continente2 = df_continenti[df_continenti["continent"] == continente2]["total_cases"].mean()
percentuale_continente1 = (total_cases_continente1 / total_cases_world) * 100
percentuale_continente2 = (total_cases_continente2 / total_cases_world) * 100

print(f"Confronto tra {continente1} e {continente2}:")
print(f"Massimo {continente1}: {massimo_continente1}, Massimo {continente2}: {massimo_continente2}")
print(f"Media {continente1}: {media_continente1}, Media {continente2}: {media_continente2}")
print(f"Percentuale rispetto al mondo {continente1}: {percentuale_continente1}%, {continente2}: {percentuale_continente2}%")
```

```python
df_italia_2022 = df[(df["location"] == "Italy") & (df["date"].str.contains("2022"))].copy()

df_italia_2022["date"] = pd.to_datetime(df_italia_2022["date"])

print(df_italia_2022.head())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_italia_2022["date"], df_italia_2022["total_cases"])
plt.title("Evoluzione dei casi totali in Italia nel 2022")
plt.xlabel("Data")
plt.ylabel("Casi totali")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_italia_2022["date"], df_italia_2022["new_cases"])
plt.title("Numero di nuovi casi giornalieri in Italia nel 2022")
plt.xlabel("Data")
plt.ylabel("Nuovi casi giornalieri")
plt.grid(True)
plt.show()
```

```python
df_italia_2022["cumulative_new_cases"] = df_italia_2022["new_cases"].cumsum()

plt.figure(figsize=(10, 6))
plt.plot(df_italia_2022["date"], df_italia_2022["cumulative_new_cases"])
plt.title("Somma cumulativa dei nuovi casi in Italia nel 2022")
plt.xlabel("Data")
plt.ylabel("Somma cumulativa dei nuovi casi")
plt.grid(True)
plt.show()
```

```python
df_icu = df[(df["location"].isin(["Italy", "Germany", "France"])) & 
            (df["date"] >= "2022-05-01") & 
            (df["date"] <= "2023-04-30")]

plt.figure(figsize=(10, 6))
df_icu.boxplot(column="icu_patients", by="location", grid=False)
plt.title("Numero di pazienti in terapia intensiva (ICU) da Maggio 2022 ad Aprile 2023")
plt.suptitle("")
plt.xlabel("Nazione")
plt.ylabel("Pazienti in terapia intensiva")
plt.show()
```

```python
df_hosp_2023 = df[(df["location"].isin(["Italy", "Germany", "France", "Spain"])) & 
                  (df["date"].str.contains("2023"))]
print(df_hosp_2023)
```

```python
# i valori nulli potremmo intrpolarli:
df_hosp_2023["hosp_patients"].interpolate(method='linear', inplace=True)

valori_nulli_post = df_hosp_2023["hosp_patients"].isnull().sum()
print(f"Numero di valori nulli dopo l'interpolazione: {valori_nulli_post}")

df_hosp_2023_sum = df_hosp_2023.groupby("location")["hosp_patients"].sum()

print(df_hosp_2023_sum)

df_hosp_2023_sum.plot(kind="bar", figsize=(10, 6))
plt.title("Somma dei pazienti ospitalizzati nel 2023")
plt.xlabel("Nazione")
plt.ylabel("Pazienti ospitalizzati")
plt.grid(True)
plt.show()
```

```python
# o potremmo fare una media
media_hosp = df_hosp_2023["hosp_patients"].mean()
df_hosp_2023["hosp_patients"].fillna(media_hosp, inplace=True)

valori_nulli_post = df_hosp_2023["hosp_patients"].isnull().sum()
print(f"Numero di valori nulli dopo la sostituzione: {valori_nulli_post}")

df_hosp_2023_sum = df_hosp_2023.groupby("location")["hosp_patients"].sum()

print(df_hosp_2023_sum)

df_hosp_2023_sum.plot(kind="bar", figsize=(10, 6))
plt.title("Somma dei pazienti ospitalizzati nel 2023")
plt.xlabel("Nazione")
plt.ylabel("Pazienti ospitalizzati")
plt.grid(True)
plt.show()
```
